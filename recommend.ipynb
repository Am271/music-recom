```python
data = pd.read_csv('spotlistr-exported-playlist (5).csv')
data1 = pd.read_csv('spotlistr-exported-playlist (6).csv', delimiter=';')
data2 = pd.read_csv('spotlistr-exported-playlist (7).csv', delimiter=';')
data3 = pd.read_csv('spotlistr-exported-playlist (8).csv', delimiter=';')
data4 = pd.read_csv('spotlistr-exported-playlist (9).csv', delimiter=';')
data5 = pd.read_csv('spotlistr-exported-playlist (10).csv', delimiter=';')
data.head()
```

```python
data1.columns = ['Track Name', 'Popularity']
data2.columns = ['Track Name', 'Popularity']
data3.columns = ['Track Name', 'Popularity']
data4.columns = ['Track Name', 'Popularity']
data5.columns = ['Track Name', 'Popularity']
df = pd.concat([data, data1, data2, data3, data4, data5])
```

```python
df['Popularity'] = df['Popularity'].apply(lambda x: (x - df['Popularity'].min()) / (df['Popularity'].max() - df['Popularity'].min()))
df['Track Name'] = df['Track Name'].str.rstrip()
df = df.drop_duplicates()
len(df)
```

```python
# Define the possible actions (music genres or songs)
actions = df['Track Name'].tolist()
print(actions)
# actions = set(actions)
actions = tuple(actions)
len(actions)
```

```python
# Initialize the Q-table with zeros
# Each row corresponds to a state, and each column corresponds to an action
Q = np.random.randn(len(actions), len(actions))
```

```python
R = np.zeros_like(Q)
R.shape
```

```python
for i in range(len(data4['Track Name']) - 1):
    # print()
    start_i = actions.index(data4['Track Name'][i].strip())
    end_i = actions.index(data4['Track Name'][i + 1].strip())
    print(start_i, end_i)

    R[start_i][end_i] = 1
    print(R[start_i][end_i])
```

```python
# Define the hyperparameters
gamma = 0.8
alpha = 0.1
epsilon = 0.5
```

```python
# Define the function to select the next action
def select_action(state):
    if np.random.rand() < epsilon:
        # Select a random action
        action = np.random.choice(np.arange(len(R)), 1)[0]
    else:
        # Select the action with the highest Q-value
        action = np.argmax(Q[state])
    return action
```

```python
# Run the Q-learning algorithm
num_episodes = 100
for episode in range(num_episodes):
    # Initialize the starting state
    state = np.random.choice(np.arange(len(R)), 1)[0]
    
    # Loop through the episode
    while True:
        # Select the next action
        action = select_action(state)
        
        # Take the action and observe the next state and reward
        next_state = action
        if action >= len(actions): print(actions)

        # pop = df.loc[df['Track Name'] == action, 'Popularity'].iloc[0] / 0.7
        # print(actions[action])
        # Get the user's feedback on the previous recommendation (0 for no, 1 for yes)
        feedback = int(input("Did you like the last recommendation? (0 for no, 1 for yes) "))
        reward = R[state][action] * feedback
        # if action in df['Track Name']:
        #     reward = R[state, action]
        # else:
        # reward = R[state, action]
        
        # Update the Q-value for the current state-action pair
        Q[state, action] = (1 - alpha) * Q[state, action] + alpha * (reward + gamma * np.max(Q[next_state]))
        
        # Update the current state
        state = next_state
        
        # Check if the episode is done
        if state == len(actions) - 1:
            break
```

```python
def suggest(song):
    sorted_index_array = np.argsort(Q[actions.index(song)])
    returnList = []
    for i in sorted_index_array[-4 : ]:
        returnList.append(actions[i])
    return returnList
```
